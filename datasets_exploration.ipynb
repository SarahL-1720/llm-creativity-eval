```python
import pandas as pd
```

```python
df_conversations = pd.read_parquet('conversations.parquet',engine='fastparquet')
df_reactions_full = pd.read_parquet('reactions.parquet', engine='pyarrow')
df_votes = pd.read_parquet('votes.parquet',engine='fastparquet')
```

```python
print(df_conversations.columns)
print('\n')
print(df_reactions_full.columns)
print('\n')
print(df_votes.columns)
```

```python
df_merged = pd.merge(df_reactions_full, df_conversations, on='conversation_pair_id', how='left',suffixes=('', '_right'))
print(len(df_merged))
print(len(df_reactions_full))
print(len(df_conversations))
print(df_merged.columns)
```

```python
test_conv_id = df_reactions_full.sample(1).iloc[0]["conversation_pair_id"]
mask_conv = df_conversations["conversation_pair_id"] == test_conv_id
mask_reac = df_reactions_full["conversation_pair_id"] == test_conv_id

print(len(df_reactions_full[mask_reac]))
print(len(df_conversations[mask_conv]))

print(df_reactions_full[mask_reac]['visitor_id'])
print("\n")
print(df_conversations[mask_conv]['visitor_id'],"\n")
```
